```python
#Right  diagonal traversal of Binary search tree

class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None


def insert(root,data):
    if root is None:
        return Node(data)
    if root.data == data:
        return data
    elif root.data < data:
        root.right = insert(root.right,data)
    else:
        root.left = insert(root.left,data)
    return root

def diagonalPrintUtil(root, d, diagonalPrintMap):
     
    if root is None:
        return

    try :
        diagonalPrintMap[d].append(root.data)
    except KeyError:
        diagonalPrintMap[d] = [root.data]
        
    diagonalPrintUtil(root.left,d+1, diagonalPrintMap)
    diagonalPrintUtil(root.right,d, diagonalPrintMap)
 
 
def diagonalPrint(root):
 
    diagonalPrintMap = dict()
    diagonalPrintUtil(root, 0, diagonalPrintMap) 
    print ("Right Diagonal Traversal of Binary Tree : ")
    for i in diagonalPrintMap:
        for j in diagonalPrintMap[i]:
            print (j,end=" ")
        print()
        
        
input_list = [8,3,10,1,6,14,13,4,7]
root = Node(input_list[0])
for e in input_list[1:]:
    insert(root,e)

diagonalPrint(root)
```

```python
#print vertical order of BST (extension)

class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None


def insert(root, data):
    if root is None:
        return Node(data)
    if root.data == data:
        return data
    elif root.data < data:
        root.right = insert(root.right, data)
    else:
        root.left = insert(root.left, data)
    return root


def inorder(root):
    if root:
        inorder(root.left)
        print(root.data, end=" ")
        inorder(root.right)


def verticalorder(root, hd, m):
    if root is None:
        return
    try:
        m[hd].append(root.data)
    except:
        m[hd] = [root.data]

    verticalorder(root.left, hd - 1, m)
    verticalorder(root.right, hd + 1, m)


def printvertical(root):
    m = dict()
    verticalorder(root, 0, m)
    for index, value in enumerate(sorted(m)):
        for i in m[value]:
            print(i, end=" ")
        print()


root = Node(8)
l_ = [3, 10, 1, 6, 14, 13, 4, 7]

for x in l_:
    insert(root, x)

print("vertical order traversal is")
printvertical(root)
```
