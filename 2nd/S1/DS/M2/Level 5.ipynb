```python
class Node:
    def __init__(self,data):
        self.data = data
        self.left = None
        self.right = None

def insert(root,data):
    if root is None:
        return Node(data)
    if root.data == data:
        return data
    elif root.data < data:
        root.right = insert(root.right,data)
    else:
        root.left = insert(root.left,data)
    return root

# def insert(root, data):
#     newnode = Node(data)
#     x = root
#     y = None
#     while (x != None):
#         y = x
#         if (data < x.data):
#             x = x.left
#         else:
#             x = x.right
#     if (y == None):
#         y = newnode
#     elif (data < y.data):
#         y.left = newnode
#     else:
#         y.right = newnode
#     return y

def inorder(root):
    if root:
        inorder(root.left)
        print(root.data,end = " ")
        inorder(root.right)

def preorder(root):
    if root:
        print(root.data,end = " ")
        preorder(root.left)
        preorder(root.right)

def postorder(root):
    if root:
        preorder(root.left)
        preorder(root.right)
        print(root.data,end = " ")
        
        
def levelorder(root):
    queue = []
    queue.append(root)
    while(len(queue) != 0):
        top = queue.pop(0)
        print(top.data,end = " ")
        if(top.left != None):
            queue.append(top.left)
        if(top.right != None):
            queue.append(top.right)
        
def search(root,data):
    if root is None:
        return root
    if root.data == data:
        return root.data
    elif root.data < data:
        return search(root.right,data)
    return search(root.left,data)


r = Node(100)
insert(r,50)
insert(r,20)
insert(r,30)
insert(r,10)
insert(r,40)
insert(r,80)

print("inorder ")
inorder(r)

print("\n")
print("preorder")
preorder(r)

print("\n")
print("post order")
postorder(r)

print('\n')
print("levelorder")
levelorder(r)

print('\n')
print("Search")
if search(r,40):
    print("found")
else:
    print("not found")
```
